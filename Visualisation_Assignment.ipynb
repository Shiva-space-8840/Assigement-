```python
#4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different categories.
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
np.random.seed(42)

categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = {
    'Category A': np.random.normal(20, 5, 100),
    'Category B': np.random.normal(35, 10, 100),
    'Category C': np.random.normal(50, 7, 100),
    'Category D': np.random.normal(25, 3, 100)
}

data = []
for category in categories:
    data.append(values[category])
import pandas as pd

df = pd.DataFrame({
    'Category': np.repeat(categories, 100),
    'Value': np.concatenate(data)
})
plt.figure(figsize=(8, 6))
sns.boxplot(x='Category', y='Value', data=df, palette='Set2')
plt.title('Distribution of Numerical Values Across Different Categories')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

```python
#5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
np.random.seed(42)
n_samples = 100
X1 = np.random.normal(0, 1, n_samples)
X2 = 0.8 * X1 + np.random.normal(0, 0.5, n_samples)
X3 = 0.5 * X1 + 0.5 * X2 + np.random.normal(0, 0.5, n_samples)
data = pd.DataFrame({'Feature 1': X1, 'Feature 2': X2, 'Feature 3': X3})
corr_matrix = data.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, cbar=True)
plt.title('Correlation Matrix of Features')
plt.show()
```

```python
#1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
from bokeh.plotting import figure, show
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)', width=800, height=400)
p.line(x, y, legend_label="sin(x)", line_width=2, color="blue")
show(p)
```

```python
#2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.
#2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.
from bokeh.plotting import figure, show
import numpy as np
import pandas as pd
np.random.seed(10)
x = np.random.rand(100) * 10
y = np.random.rand(100) * 10
sizes = np.random.rand(100) * 20 + 5
colors = np.random.rand(100)
data = pd.DataFrame({'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

p = figure(title="Scatter Plot with Varying Sizes and Colors",x_axis_label='X values',y_axis_label='Y values',width=800,height=400)
p.scatter('x', 'y', size='sizes', color='colors', source=data, legend_field='colors',fill_alpha=0.6, line_color=None)
show(p)
```

```python
#3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
from bokeh.plotting import figure, show
from bokeh.io import output_file
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]
p = figure(x_range=fruits, title="Fruit Counts", toolbar_location=None, plot_height=300, plot_width=400)
p.vbar(x=fruits, top=counts, width=0.5, color="blue", legend_label="Fruits", alpha=0.6)
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.legend.orientation = "horizontal"
p.legend.location = "top_center"
p.xaxis.axis_label = "Fruit"
p.yaxis.axis_label = "Count"
output_file("fruit_counts_bar_chart.html")
show(p)
```
